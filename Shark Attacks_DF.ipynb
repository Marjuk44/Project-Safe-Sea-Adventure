```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

url = 'https://www.sharkattackfile.net/spreadsheets/GSAF5.xls'

shark_attack_df1 = pd.read_excel(url)
shark_attack_df1
```

```python
shark_attack_df1.info()
```

```python
columns_to_remove = shark_attack_df1.columns[15:23]
shark_attack_df1_cleaned = shark_attack_df1.drop(columns=columns_to_remove)
shark_attack_df1_cleaned.shape
```

```python
#Keep data from 1980 to 2024
start_year = 1980.0
end_year = 2024.0
shark_attack_df1_cleaned = shark_attack_df1_cleaned[(shark_attack_df1_cleaned['Year']>=start_year) & (shark_attack_df1_cleaned['Year']<=end_year)]

shark_attack_df1_cleaned.shape
```

```python
shark_attack_df1_cleaned
```

```python
# 1. Handling Missing Values: Drop rows with missing 'Date' or 'Location'
shark_attack_df1_cleaned = shark_attack_df1_cleaned.dropna(subset=['Date', 'Location'])
print(shark_attack_df1_cleaned.shape)
```

```python
shark_attack_df1_cleaned
```

```python
# 2. Removing Duplicates: Drop duplicate rows
shark_attack_df1_cleaned = shark_attack_df1_cleaned.drop_duplicates()
print(shark_attack_df1_cleaned.shape)
```

```python
# 3. Correcting Data Types: Convert 'Date' to datetime
shark_attack_df1_cleaned['Date'] = pd.to_datetime(shark_attack_df1_cleaned['Date'], errors='coerce')
print(shark_attack_df1_cleaned.info())
```

```python
# 4. Standardizing Data Entries: Normalize attack types
attack_type_corrections = {
    'Invalid': 'Provoked', 'Watercraft': 'Provoked', '?': 'Provoked',
    'Questionable': 'Provoked', 'Sea Disaster': 'Provoked', 'Unconfirmed': 'Provoked',
    'Under investigation': 'Provoked', 'Unverified': 'Provoked', ' Provoked': 'Provoked'
}
shark_attack_df1_cleaned['Type'] = shark_attack_df1_cleaned['Type'].replace(attack_type_corrections)

# Normalize names
name_corrections = {
    'male': 'John Doe', 'female': 'John Doe', 'boy': 'John Doe', '2 male': 'John Doe',
    'boat': 'John Doe', 'Unidentified': 'John Doe', 'girl': 'John Doe', 'Anonymous': 'John Doe',
    'child': 'John Doe', 'A male from Muitoa': 'John Doe', '2 males': 'John Doe'
}
shark_attack_df1_cleaned['Name'] = shark_attack_df1_cleaned['Name'].replace(name_corrections)

# Normalize sex categories
sex_corrections = {'M ': 'M', ' M': 'M', 'N': 'M', 'Mx2': 'M', 'lli': 'M'}
shark_attack_df1_cleaned['Sex'] = shark_attack_df1_cleaned['Sex'].replace(sex_corrections)
print(shark_attack_df1_cleaned['Sex'].value_counts())
```

```python
# 5. Converting Dates: Ensure 'Year' is derived from 'Date'
shark_attack_df1_cleaned['Year'] = shark_attack_df1_cleaned['Date'].dt.year
shark_attack_df1_cleaned.head(101)
```

```python
# Summary Statistics
shark_attack_df1_cleaned.describe(include='all')
```

```python
# Visualize distribution of shark attacks by year
plt.figure(figsize=(12, 8))
sns.countplot(data=shark_attack_df1_cleaned, x='Year')
plt.title('Distribution of Shark Attacks by Year (1980-2024)')
plt.xticks(rotation=90)
plt.show()
```

```python
# Visualize top 5 beaches to avoid
top_beaches_to_avoid = shark_attack_df1_cleaned['Location'].value_counts().nlargest(5)
print("Top 5 Beaches to Avoid:\n", top_beaches_to_avoid)
```

```python
# Visualize top 5 beaches to visit
top_beaches_to_visit = shark_attack_df1_cleaned['Location'].value_counts().nsmallest(5)
print("Top 5 Beaches to Visit:\n", top_beaches_to_visit)
```

```python
# Visualization of attack types
plt.figure(figsize=(12, 8))
sns.countplot(data=shark_attack_df1_cleaned, x='Type')
plt.title('Distribution of Shark Attack Types')
plt.xticks(rotation=90)
plt.show()
```

```python
# Chart to show the time of the day when the attack happened
plt.figure(figsize=(12, 8))
sns.countplot(data=shark_attack_df1_cleaned, x='Time')
plt.title('Shark Attacks by Time of Day')
plt.xticks(rotation=90)
plt.show()
```

```python
# Save the cleaned data for further analysis
shark_attack_df1_cleaned.to_csv("shark_attack_cleaned.csv", index=False)
shark_attack_df1_cleaned.head(101)
```
